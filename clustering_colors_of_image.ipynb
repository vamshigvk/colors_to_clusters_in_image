```python
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread,imshow
from IPython.display import display
from sklearn.cluster import KMeans
```

```python
img0 = imread('https://upload.wikimedia.org/wikipedia/commons/3/34/Make_a_pancake.jpg')
img1 = imread('http://veggiefoodrecipes.com/wp-content/uploads/2016/05/lentil-samosa-recipe-01.jpg')
```

```python
imshow(img1)
```

```python
imshow(img0)
```

```python
img0.shape
```

```python
img1.shape
```

```python
X0 = img0.reshape(-1,3)/255
X0.shape
```

```python
X1 = img1.reshape(-1,3)/255
X1.shape
```

```python
kmeans0 = KMeans(n_clusters=3)
kmeans1 = KMeans(n_clusters=3)
```

```python
kmeans0.fit(X0)
kmeans1.fit(X1)
```

```python
kmeans0.cluster_centers_
```

```python
kmeans1.cluster_centers_
```

```python
kmeans0.labels_
```

```python
kmeans1.labels_
```

```python
new_img0 = kmeans0.cluster_centers_[kmeans0.labels_]
new_img1 = kmeans1.cluster_centers_[kmeans1.labels_]
```

```python
image1 = new_img1.reshape(525, 700, 3)
image0 = new_img0.reshape(480, 640, 3)
```

```python
imshow(image0)
```
